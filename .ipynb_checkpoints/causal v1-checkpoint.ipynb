```python
%%time
# load daily gmv, spend, cm for each leaf category
ts = pd.read_sql("""
                select dt, ramp, a.leaf_category, 
                            sum(gmv_w_cancel_est) gmv_w_cancel_est,
                            sum(spend) spend,
                            sum(cm) cm
                from mp.order_item_cm_metrics a
                left join wandajuan.llcm_leaf_split b 
                on a.leaf_category = b.leaf_category 
                where dt >= '2019-01-01'
                group by 1, 2, 3
                """, conn)
ts.head()
```

```python
ts.shape
```

```python
ts['dt'] = pd.to_datetime(ts['dt'])
ts['year'] = ts['dt'].dt.year
ts['month'] = ts['dt'].dt.to_period('M')
ts['week'] = ts['dt'].dt.to_period('W')
ts.head()
```

```python
ts.groupby(['week', 'ramp'])['gmv_w_cancel_est'].mean().reset_index().\
        pivot('week', 'ramp', 'gmv_w_cancel_est').plot(figsize=(16, 6))
plt.vlines(x=pd.to_datetime('2021-10-22'), ymin=500, ymax=2000, linestyle=':', label='llcm ramp1')
plt.legend()
```

```python
# zooming into 2021
ts_2021 = ts.query('year==2021')
ts_2021.groupby(['week', 'ramp'])['gmv_w_cancel_est'].mean().reset_index().\
        pivot('week', 'ramp', 'gmv_w_cancel_est').plot(figsize=(16, 6))
plt.vlines(x=pd.to_datetime('2021-10-22'), ymin=500, ymax=2000, linestyle=":", lw=2, label='llcm ramp1')
plt.ylabel('gmv_w_cancel_est')
plt.title('2021 gmv trend')
plt.legend()
```

# T1

```python
# filter for ramp is not null in 2021 and exclude dining tables
ts_T1 = ts.query('year==2021 and ramp == ramp and leaf_category != "dining tables"') 
ts_T1
ts_T1.assign(T1 = np.where(ts_T1['ramp']=='T1', 'T1', 'Other'))\
        .groupby(['week', 'T1'])['gmv_w_cancel_est'].mean().reset_index().\
        pivot('week', 'T1', 'gmv_w_cancel_est').plot(figsize=(16, 6))
plt.vlines(x=pd.to_datetime('2021-10-22'), ymin=500, ymax=2000, linestyle=":", lw=2, label='llcm ramp1')
plt.ylabel('gmv_w_cancel_est')
plt.title('2021 gmv trend')
plt.legend()
```

```python
# filter for ramp is not null in 2021 and exclude dining tables
ts_T1 = ts.query('year==2021 and ramp == ramp and leaf_category != "dining tables"') 
ts_T1
ts_T1.assign(T1 = np.where(ts_T1['ramp']=='T1', 'T1', 'Other'))\
        .groupby(['week', 'T1'])['cm'].mean().reset_index().\
        pivot('week', 'T1', 'cm').plot(figsize=(16, 6))
plt.vlines(x=pd.to_datetime('2021-10-22'), ymin=0, ymax=250, linestyle=":", lw=2, label='llcm ramp1')
plt.ylabel('cm')
plt.title('2021 cm trend')
plt.legend()
```

# model on preperiod

```python
preperiod = ts_T1.query('dt < "2021-10-22"')\
        .groupby(['week', 'ramp'])['gmv_w_cancel_est'].mean().reset_index()\
        .pivot(index='ramp', columns='week').T

preperiod
```

```python
y_train = preperiod['T1'].values
X_train = preperiod.drop(columns='T1').values
y_train.shape, X_train.shape
```

```python
from sklearn.linear_model import Lasso
clf = Lasso(fit_intercept=False).fit(X_train, y_train)
weights_lr = clf.coef_
weights_lr
```

```python
allperiod = ts_T1\
        .groupby(['week', 'ramp'])['gmv_w_cancel_est'].mean().reset_index()\
        .pivot(index='ramp', columns='week').T

y_true = allperiod['T1'].values
X = allperiod.drop(columns='T1').values
y_pred = clf.predict(X)
```

```python
weeks = pd.Series([ind[1] for ind in allperiod.index])
weeks.shape
```

```python
y_true.shape, y_pred.shape
```

```python
# construct all period
y_all = pd.DataFrame(index=weeks, columns=['T1 actual', 'T1 predicted'], data=np.array([y_true, y_pred]).T)
y_all.plot(figsize=(16, 6))
plt.vlines(x=pd.to_datetime('2021-10-22'), ymin=0, ymax=1700, linestyle=":", lw=2, label='llcm ramp1')
plt.ylabel('gmv')
plt.title('2021 gmv trend')
plt.legend()
```
